```python
#To check whether the given number is palindrome or not
def is_palindrome(S):
  S=S.replace("","").lower()
  return S==S[::-1]

user_input = input("Enter a string which to check if its a palindrome :")

#call the function and check if the input is a palindrome
if is_palindrome(user_input):
  print(f"{user_input} is a palindrome.")
else:
  print(f"{user_input} is not a palindrome.")
```
